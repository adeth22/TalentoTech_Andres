```python
# Importar librerías
import requests
import pandas as pd
import io
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Leer archivo excel con pandas de una carpeta compartida.
url = 'https://drive.google.com/uc?export=download&id=1lVqih5wnGrhzDNnHLlabe1f_WiGCcj7P'

# Stream the response content directly into BytesIO
response = requests.get(url, stream=True)
response.raise_for_status()

# Use BytesIO to handle the binary data
excel_data = io.BytesIO(response.content)

# Read the excel file from the BytesIO object
df_ventas = pd.read_excel(excel_data)

df_ventas # Visualizar el conjunto de datos
```

```python
# Ejercicio 1: Mostrar las primeras filas
print(f"Ejercicio 1: Mostrar las primeras filas" )
df_ventas.head(5)
```

```python
# Ejercicio 2: Calcular el total de ventas por región
print(f"Ejercicio 2: Calcular el total de ventas por región" )
ventas_por_region = df_ventas.groupby('Region')['Ventas'].sum()
ventas_por_region
```

```python
# Ejercicio 3 Filtrar el Dataframe para mostrar solo las ventas realizadas por David
print("Ejercicio 3: Filtrar el Dataframe para mostrar solo las ventas realizadas por David")
ventas_david = df_ventas[df_ventas["Vendedor"]=="David"]
ventas_david
```

```python
# Ejercicio 4: Calcular el promedio de ventas por región
print("Ejercicio 4: Calcular el promedio de ventas por región")
ventas_por_region = df_ventas.groupby("Region")["Ventas"].mean()
ventas_por_region
```

```python
# Ejercicio 5: Calcular el promedio de ventas por región con dos decimales
print("Ejercicio 5: Calcular el promedio de ventas por región con dos decimales")
ventas_por_region = ventas_por_region.round(2)
ventas_por_region
```

```python
# Ejercicio 6: Ordenar el Dataframe por ventas en orden descendentes
print("Ejercicio 6: Ordenar el Dataframe por ventas en orden descendentes")
ventas_ordenado = df_ventas.sort_values(by="Ventas",ascending=False)
ventas_ordenado
```

```python
# Ejercicio 7: Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor
print("Ejercicio 7: Filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor")
df_filtrado = df_ventas[df_ventas["Ventas"]>10000].sort_values(by="Ventas", ascending=False)
df_filtrado
```

```python
# Ejercicio 8: Eliminar la columna "Vendedor"
print("Ejercicio 8: Eliminar la columna Vendedor")
df_no_vendedor = df_ventas.drop(columns=["Vendedor"])
df_no_vendedor
```

```python
# Pregunta 9: ¿Cuál es el top 3 de los producto más vendido?
ventas_mas_vendido = df_ventas.groupby('Producto')['Ventas'].sum().sort_values(ascending=False)
ventas_mas_vendido.head(3)
```
